## Addressing Missing Data/ Default Values

### As you cleaned your data, you may have observed missing data/ default values. Decide what to do with these values and explain your decision-making process.

Here are some methods to deal with missing values:
- Drop records with missing values
    - Check for biases induced by missing values - there may be trends in why something is missing
    - This requires domain knowledge to check for induced bias
- Imputation: Inferring missing values
    - Replace with average value (closest related subgroup mean)
    - Replace with a random value from the closest subgroup(Hot deck imputation) 
- Numerical: 
    - mean, mode, median
- Categorical
    - Mode, new category
- Directly model missing values during future analysis

## Handling outliers

### Now let's handle any outliers. Same process as before!

Here are some methods to deal with handling outliers:
- Drop records with outliers
    - Check for biases induced by missing values - there may be trends in why something is missing
    - This requires domain knowledge to check for induced bias
- Replace, as with Inferring missing values
    - Typically with min or max
- Binning
    - E.g. greater than 5

```python
stats = breast_cancer.describe().transpose()
IQR = stats['75%'] - stats['25%']
stats['75% + 1.5IQR'] = (1.5 * IQR) + stats['75%']
stats['25% - 1.5IQR'] = stats['25%'] - (1.5 * IQR)
stats

q_high = breast_cancer["radius_mean"].quantile(0.99)
q_low = breast_cancer["radius_mean"].quantile(0.01)

breast_cancer_filtered = breast_cancer[(breast_cancer["radius_mean"] < q_high) & (breast_cancer["radius_mean"] > q_low)]
breast_cancer_filtered

for col in breast_cancer.columns[2:]:
    q_high = breast_cancer[col].quantile(0.99)
    q_low = breast_cancer[col].quantile(0.01)
    breast_cancer_filtered = breast_cancer[(breast_cancer[col] < q_high) & (breast_cancer[col] > q_low)]
breast_cancer_filtered
```

```python
#I first calculated the IQR and added a column to my stats table which was obtained by using the .describe() and .transpose() methods.
# I then created an upper bound and lower bound by doing the 3rd quartile + 1.5 * IQR, and 1st quartile - 1.5 *IQR and also added these to my stats table
# To actually filter out values, I used a for loop to iterate through all numerical columns and find the higher quantile of .99 and lower quantile of .01 and filter out the values in each column that don't fall in between these quantiles
```

```python
#Visualization 1

size = len(breast_cancer_filtered['texture_mean'])


area = np.pi * (15 * np.random.rand( size ))**2
colors = np.random.rand( size )


plt.xlabel("texture mean")
plt.ylabel("radius mean") 
plt.scatter(breast_cancer_filtered['texture_mean'], breast_cancer_filtered['radius_mean'], s=area, c=colors, alpha=0.5);


# I did a scatter plot with special sizes to indicate the amount of datapoints a certain area has, and special colors to differentiate each point and just make it easier to visualize
# There seems to be a slight positive correlation between texture mean and radius mean
```

```python
#Visualization 2

sns.violinplot(data = breast_cancer_filtered, y = 'radius_mean')

# This is a violin plot of the radius mean column. It indicates that the distribution is most likely unimodal as there is a lot of datapoints located around the value of 12.5
```

```python
#Visualization 3

sns.lmplot(data = breast_cancer_filtered, x = 'radius_mean', y = 'texture_mean', fit_reg = True)

# Here I plotted a scatterplot of radius_mean vs texture mean with a regression line. The regression line indicates a slight positive correlation between the 2.
```

```python
#Visualization 4

sns.boxplot(data = breast_cancer_filtered, y = 'radius_mean')

# Here I plotted a boxplot of the radius mean column, and it shows how the median of the data is around 13, and there seems to be outliers at the top spectrum of the data
```

```python
#Visualization 5

import seaborn as sns
sns.histplot(data = breast_cancer_filtered, x = 'radius_mean', hue = 'diagnosis', stat = 'density', kde = True)

#Here I made a histogram plot with KDE, which gives us a better idea of where the peaks of the data lie and how it is skewed. I also did an overlaying histogram to see the distributions of the malignant masses and the benign masses to see how they compare. It is evident that malignant masses have a higher radius mean than those that are benign
```
